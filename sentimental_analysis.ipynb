```python
import datasets
from transformers import pipeline
from transformers.pipelines.pt_utils import KeyDataset
from tqdm.auto import tqdm
```

```python
import pandas as pd
hdfc = pd.read_csv("HDFC_news_21.csv")
hdfc.tail()
```

```python
hdfc_rev = hdfc.iloc[::-1].reset_index(drop=True)
```

```python
hdfc_rev.tail()
```

```python
data = hdfc_rev
```

```python
from transformers import pipeline

pipe = pipeline("text-classification", model="ProsusAI/finbert", top_k = None)
lables = ["positive", "negative", 'neutral']
```

```python
results = pipe(data["headline"].tolist())
```

```python
parsed_rows = []
for rows in results :
    clean_dict = {'positive': 0.0, 'negative': 0.0, 'neutral': 0.0}

    for pred in rows:
        label = pred['label']
        score = pred['score']
        clean_dict[label] = score
        parsed_rows.append(clean_dict)

output = pd.DataFrame(parsed_rows)
    
```

```python
output['Date'] = hdfc['date']
```

```python
output = output.iloc[::3]
output.reset_index(drop=True)
```

```python
output.to_csv('sentiment.csv',index = False )
```
